Данные связаны с прямыми маркетинговыми кампаниями (телефонными звонками) португальского банковского учреждения. Цель классификации-предсказать, подпишется ли клиент на срочный депозит (переменная y).

Входные параметры:

Данные о клиенте банка:  
1 - age(возраст) (numeric)   
2 - job(тип работы) (categorical: 'admin.','blue-collar','entrepreneur','housemaid','management','retired','self-employed','services','student','technician','unemployed','unknown')   
3 - marital(семейное положение) (categorical: 'divorced','married','single','unknown'; заметка: 'divorced' означает разведенный или овдовевший)   
4 - education(образование) (categorical: 'tertiary','secondary','unknown','primary)   
5 - default(имеет ли кредит по-умолчанию) (categorical: 'no','yes','unknown')   
6 - balance(остаток средств) (numeric)  
7 - housing(имеет ли жилищный кредит) (categorical: 'no','yes','unknown')   
8 - loan(имеет ли личный кредит?) (categorical: 'no','yes','unknown')  

Данные, связанные с последним контактом с клиентом:  
9 - contact(тип связи) (categorical: 'cellular','telephone')    
10 - day(день последнего контакта) (numeric)   
11 - month(месяц последнего контакта) (categorical: 'jan', 'feb', 'mar', ..., 'nov', 'dec')  
12 - duration(длительность последнего разговора) (numeric).   

Другие признаки:  
13 - campaign(количество контактов с клиентом) (numeric)   
14 - pdays(количество дней прошедших с последнего контакта с клиентом) (numeric; 999 значит, что с клиентом ранее не связывались)   
15 - previous(количество предыдущих контактов с клиентом) (numeric)   
16 - poutcome(итоги предыдущей маркетинговой компании) (categorical: 'failure','nonexistent','success')  

Выходные параметры:   
17 - y(подписал ли клиент срочный вклад?) (binary: 'yes','no')

Рассмотрим результат обучения для разных гиперпараметров alpha

```python
alpha_arr = np.logspace(-3, 2, 21)
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (100,), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(alpha_arr, train_err, 'b-o', label = 'test')
plt.semilogx(alpha_arr, test_err, 'r-o', label = 'train')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
```

Минимальные значения ошибки. Видно, что эксперимент с гиперпараметров не сильно улучшил результаты

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

Оптимальное значение гиперпараметра

```python
alpha_opt = alpha_arr[test_err == min_test_err]
print(alpha_opt)
```

Повтор обучения

```python
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (100,),
                          solver = 'lbfgs', activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)

y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print(np.mean(y_train != y_train_pred))
print(np.mean(y_test != y_test_pred))
```

Матрица рассогласования

Эксперимент с изменением размера скрытого слоя

```python
size_arr = np.logspace(0, 4, 11)
test_err = []
train_err = []
for size in size_arr:
    mlp_model = MLPClassifier(hidden_layer_sizes = (int(size),), solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
    
plt.semilogx(size_arr, train_err, 'b-o', label = 'test')
plt.semilogx(size_arr, test_err, 'r-o', label = 'train')
plt.xlim([np.max(size_arr), np.min(size_arr)])
plt.title('Error vs. numbers of neirons')
plt.xlabel('number')
plt.ylabel('error')
plt.legend()
pass
```

Перепутаны надписи. При размере слоя около 50 наблюдается наибольшее переобучение. Ошибки для тестовой выборки колеблются от 0.092 до 0.1

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

```python
size_opt = size_arr[test_err == min_test_err]
print(size_opt)
```

В результате получены значения оптимального размера слоя size=100 и оптимального гиперпараметра alpha=1.

Random forest

```python
from sklearn import ensemble
size_arr = np.logspace(0, 4, 11)
test_err = []
train_err = []
oob_err = []
for size in size_arr:
    rf_model = ensemble.RandomForestClassifier(n_estimators = int(size), oob_score=True)
    rf_model.fit(X_train, y_train)

    y_train_pred = rf_model.predict(X_train)
    y_test_pred = rf_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
    oob_err.append(1-rf_model.oob_score)
    
plt.semilogx(size_arr, train_err, 'b-o', label = 'train')
plt.semilogx(size_arr, test_err, 'r-o', label = 'test')
plt.semilogx(size_arr, oob_err, 'g-o', label = 'oob_err')
plt.xlim([np.max(size_arr), np.min(size_arr)])
plt.title('Error vs. numbers of trees')
plt.xlabel('tree')
plt.ylabel('error')
plt.legend()
pass
```

Ошибка на обучающей выборке равная 0 говорит о переобучении, при этом ошибка на тестовой выборке остается большой.
