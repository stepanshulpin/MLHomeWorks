Данные связаны с прямыми маркетинговыми кампаниями (телефонными звонками) португальского банковского учреждения. Цель классификации-предсказать, подпишется ли клиент на срочный депозит (переменная y).

Входные параметры:

Данные о клиенте банка:  
1 - age(возраст) (numeric)   
2 - job(тип работы) (categorical: 'admin.','blue-collar','entrepreneur','housemaid','management','retired','self-employed','services','student','technician','unemployed','unknown')   
3 - marital(семейное положение) (categorical: 'divorced','married','single','unknown'; заметка: 'divorced' означает разведенный или овдовевший)   
4 - education(образование) (categorical: 'tertiary','secondary','unknown','primary)   
5 - default(имеет ли кредит по-умолчанию) (categorical: 'no','yes','unknown')   
6 - balance(остаток средств) (numeric)  
7 - housing(имеет ли жилищный кредит) (categorical: 'no','yes','unknown')   
8 - loan(имеет ли личный кредит?) (categorical: 'no','yes','unknown')  

Данные, связанные с последним контактом с клиентом:  
9 - contact(тип связи) (categorical: 'cellular','telephone')    
10 - day(день последнего контакта) (numeric)   
11 - month(месяц последнего контакта) (categorical: 'jan', 'feb', 'mar', ..., 'nov', 'dec')  
12 - duration(длительность последнего разговора) (numeric).   

Другие признаки:  
13 - campaign(количество контактов с клиентом) (numeric)   
14 - pdays(количество дней прошедших с последнего контакта с клиентом) (numeric; 999 значит, что с клиентом ранее не связывались)   
15 - previous(количество предыдущих контактов с клиентом) (numeric)   
16 - poutcome(итоги предыдущей маркетинговой компании) (categorical: 'failure','nonexistent','success')  

Выходные параметры:   
17 - y(подписал ли клиент срочный вклад?) (binary: 'yes','no')

```python
data_describe = data.describe(include = [object])
for c in categorical_columns:
    data[c] = data[c].fillna(data_describe[c]['top'])
```

```python
binary_columns    = [c for c in categorical_columns if data_describe[c]['unique'] == 2]
nonbinary_columns = [c for c in categorical_columns if data_describe[c]['unique'] > 2]
print(binary_columns, nonbinary_columns)
```

Создание на основе категориальных небинарных признаков дополнительных признаков, включающих все возможные состояния

```python
data['marital'].unique()
```

```python
data_nonbinary = pd.get_dummies(data[nonbinary_columns])
print(data_nonbinary.columns)
```

В итоге все признаки переведены в числовые

```python
data_numerical = data[numerical_columns]
data_numerical.describe()
```

Нормировка

```python
data_numerical = (data_numerical - data_numerical.mean(axis = 0))/data_numerical.std(axis = 0)
```

```python
data_numerical.describe()
```

```python
data = pd.concat((data_numerical, data_nonbinary, data[binary_columns]), axis = 1)
print(data.shape)
#print data.columns
```

Раздедение данных на обучающую и тестовую выборки

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

Обучение с использованием метода k ближайших соседей.  
k = 10

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)
#knn.set_params(n_neighbors=10)
knn.fit(X_train, y_train)
```

```python
knn
```

Ошибки на обучающей и тестовой выборке около 10%. Причем ошибка на обучающей выборке меньше. Таким образом, наблюдается ситуация переобучения.

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

```python
err_test = 1 - knn.score(X_test, y_test)
```

```python
err_test
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```
